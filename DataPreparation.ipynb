# Number of calls per week per borrough

```
calls_consolid['week'] = calls_consolid['CallDateAndTimeStart'].dt.isocalendar().week
```

```
calls_consolid['year'] = calls_consolid['CallDateAndTimeStart'].dt.isocalendar().year
```

```
calls_consolid['year-week']  = calls_consolid.apply(lambda x : datetime.strptime(str(x['year']) + ' ' + str(x['week']) + ' 1', "%Y %W %w"), axis=1 )
```

```
def get_zero_count_df(df: pd.DataFrame) -> pd.DataFrame:
  min_date = df['year-week'].min().to_pydatetime()
  max_date = df['year-week'].max().to_pydatetime()
  
  date_weeks = pd.date_range(min_date, max_date, freq='7D').to_frame()
  date_weeks.columns=['year-week']
  date_weeks.set_index('year-week', inplace=True)

  munis = pd.Series(df['Municipalité_Nom'].unique())
  index = pd.MultiIndex.from_product([munis.values, date_weeks.index.values], names= ['Municipalité_Nom', 'year-week'] )
  df_out = pd.DataFrame(index = index).reset_index()
  df_out['zero_count'] = [0] * len(df_out)
  df_out.set_index(['Municipalité_Nom', 'year-week'], inplace=True)
  return df_out
```

```
def count_calls_per_week(df_in: pd.DataFrame) -> pd.DataFrame:
  df = df_in[['year', 'week', 'year-week', 'Municipalité_Code', 'Municipalité_Nom']].value_counts().to_frame().sort_values(by=['year-week', 'Municipalité_Nom'])
  df.columns = ['nbr_of_calls']
  df = df.reset_index()
  df = df[[ 'Municipalité_Nom','year-week', 'nbr_of_calls']]
  df.set_index(['Municipalité_Nom', 'year-week'], inplace=True)

  # print('debug')
  df_zero_count = get_zero_count_df(df_in) 
  # print(df_zero_count.groupby('Municipalité_Nom').count())
  df = df.join(df_zero_count, how='right').sort_index()
  # print(df.groupby('Municipalité_Nom').count())
  df['nbr_of_calls'] = df['nbr_of_calls'].fillna(0)
  df.drop('zero_count', axis='columns', inplace=True)
  # print(df.groupby('Municipalité_Nom').count())

  df_cumul = df.groupby(['Municipalité_Nom', 'year-week']).sum().groupby(level=[0]).transform(pd.Series.cumsum).reset_index()
  df_cumul.rename(columns={'nbr_of_calls':'cumul_nbf_of_calls'}, inplace=True)
  df_cumul.set_index(['Municipalité_Nom', 'year-week'], inplace=True)
  
  df_out = df.join(df_cumul)
  df_out = df_out.sort_index().reset_index()
  return df_out
```

```
mtl_calls =  calls_consolid[calls_consolid['Région Administrative'] == 'Montréal']
df_cumul = count_calls_per_week(mtl_calls)
df_cumul.head(40)
```

```
df_cumul.to_csv('211callsByWeek_all_dates.csv')
```

```
# Filter to start at the same time of covid dataset
recent_mtl_calls = mtl_calls[mtl_calls['year-week'] >= datetime(2020,4,6)]
df_cumul_recent = count_calls_per_week(recent_mtl_calls)
df_cumul_recent.tail(40)
```
